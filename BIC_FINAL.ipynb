```python
def sigmoid(X, weight):
    z=np.dot(X, weight)
    return 1/(1+np.exp(-z))
```

```python
def loss(h,y):
    return (-y*np.log(h)-(1-y)*np.log(1-h)).mean()
```

```python
def gradient_descent(X,h,y):
    return np.dot(X.T, (h-y))/y.shape[0]
```

```python
LeRa = 0.01
def update_weight_loss(weight, LeRa, gradient):
    return weight - LeRa * gradient_descent
```

```python
def predict_probs(X, weight):
    return sigmoid(np.dot(X, weight))
```

```python
def predict(X, weight, thres=0.5):
    return predict_probs(X, weight)>= thres
```

```python
class LogReg:
    def __init__(self, LeRa=0.01, no_of_iter=2000, fit_intercept=True, verbose=False):
        self.LeRa = LeRa
        self.no_of_iter = no_of_iter
        self.fit_intercept = fit_intercept
        self.verbose = verbose
    
    def __add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def __sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    def __loss(self, h, y):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    
    def fit(self, X, y):
        print('three')
        if self.fit_intercept:
            X = self.__add_intercept(X)
        
        # weights initialization
        self.weight = np.zeros(X.shape[1])
        print('four')
        for i in range(self.no_of_iter):
            print(i, '----------------------------------------', end='\r')
            if (i % 100) == 0:
              print(i)
            z = np.dot(X, self.weight)
            h = self.__sigmoid(z)
            gradient = np.dot(X.T, (h - y)) / y.size
            self.weight -= self.LeRa * gradient
            
            if(self.verbose == True and i % 10000 == 0):
                z = np.dot(X, self.weight)
                h = self.__sigmoid(z)
                print(f'loss: {self.__loss(h, y)} \t')
    
    def predict_prob(self, X):
        print('two')
        if self.fit_intercept:
            X = self.__add_intercept(X)
    
        return self.__sigmoid(np.dot(X, self.weight))
    
    def predict(self, X, threshold):
        print('one')
        return self.predict_prob(X) >= threshold
```

```python
def model(X, y):
    model = LogReg(LeRa=0.1, no_of_iter=300000)
    %time model.fit(X, y)
    return model
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
acc_list2=[]
```

```python
class NN_sig:
    
    def __init__(self,X, y, X_test, y_test, H_N=12, L_R=0.1, epochs=5000):
        
        #data
        self.y = y[:,None]
        self.X = X
        
        self.X_test = X_test
        self.y_test = y_test


        #parameters
        np.random.seed(4)
        self.input_nodes = len(X[0])
        self.H_N = H_N
        self.output_nodes = self.y.shape[1]
        self.L_R = L_R
        
        #init weights
        self.w1 = 2*np.random.random((self.input_nodes, self.H_N)) - 1
        self.w2 = 2*np.random.random((self.H_N, self.output_nodes)) - 1

        self.train(epochs)
        self.test()
        
    def sigmoid(self,X):
        return (1/(1+np.exp(-X)))
    
    def deriv_sigmoid(self,X):
        return X * (1 - X)
        
    def train(self, epochs):
        
        for e in range(epochs):
            
            #FORWARD PROPAGATION
            
            #hidden layer
            # W1(398,30) X(30,12)
            l1 = self.sigmoid(np.dot(self.X, self.w1))

            #output layer
            #l1(398,12) W2(12,1)
            l2 = self.sigmoid(np.dot(l1, self.w2))
        
            # BACKPROPAGATION
            
            #calculate how far off our prediciton was
            error = self.y-l2
            
            #calculate how far off each layer is
            l2_Back = error * self.deriv_sigmoid(l2)
            l1_Back = l2_Back.dot(self.w2.T) * self.deriv_sigmoid(l1)

            #update weights with our newly found error values
            self.w2 = np.add(self.w2, l1.T.dot(l2_Back) * self.L_R)
            self.w1 = np.add(self.w1, self.X.T.dot(l1_Back) * self.L_R)
        
        #print('Error:', (abs(error)).mean())
    
    def test(self):
        correct = 0
        pred_list = []
        
        #replicate feedforward network for testing
        l1 = self.sigmoid(np.dot(self.X_test, self.w1))
        l2 = self.sigmoid(np.dot(l1, self.w2))
        
        #loop through all of the outputs of layer 2
        for i in range(len(l2)):
            if l2[i] >= 0.5:
                pred = 1
            else:
                pred = 0

            if pred == self.y_test[i]:
                correct += 1
                
            pred_list.append(pred)
        print(pred_list)
        print(y_test)
        correct = np.sum(pred_list == y_test)
        print("%d out of %d predictions correct" % (correct, len(pred_list)))

        print("Test Accuracy: ", ((correct/len(y_test))*100),'%')
        acc_list2.append((correct/len(y_test))*100)

        #confusion matrix
        from sklearn.metrics import confusion_matrix
        cm = confusion_matrix(y_test, pred_list)
        sns.heatmap(cm,annot=True)
        #plt.savefig('h.png')
        plt.show()
```
